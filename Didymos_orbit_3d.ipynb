```python
from astropy.utils.data import conf
conf.remote_timeout = 10000
```

```python
#astropyより単位と時間のモジュールを呼び出す
from astropy import units as u
from astropy.time import Time

#poliastroのclassで定義されている惑星を呼び出す
from poliastro.bodies import Sun, Earth, Mars, Jupiter
from poliastro.twobody import Orbit
from poliastro.plotting import OrbitPlotter3D
from poliastro.plotting.misc import plot_solar_system
from poliastro.util import time_range, norm
```

```python
#計算する時刻
EPOCH = Time("2021-11-24 00:00:00", scale="tdb")
```

```python
#NAS/JPLよりdidymosの軌道データを呼び出して軌道を計算させる
didymos_orbit= Orbit.from_sbdb("6583")
```

```python
frame = OrbitPlotter3D()

#惑星を描写
frame.plot_body_orbit(Earth, EPOCH, label="Earth")
frame.plot_body_orbit(Mars, EPOCH,label="Mars")
frame.plot_body_orbit(Jupiter, EPOCH, label="Jupiter")
frame.plot(didymos_orbit, label="Didymos")
```
