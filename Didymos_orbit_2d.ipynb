```python
from astropy.utils.data import conf
conf.remote_timeout = 10000
```

```python
#astropyより単位と時間のモジュールを呼び出す
from astropy import units as u
from astropy.time import Time

#poliastroのclassで定義されている惑星を呼び出す
from poliastro.bodies import Sun, Earth, Mars, Jupiter
from poliastro.ephem import Ephem
from poliastro.frames import Planes
from poliastro.twobody import Orbit
from poliastro.plotting import StaticOrbitPlotter
from poliastro.plotting.misc import plot_solar_system
```

```python
#計算する時刻
EPOCH = Time("2021-11-24 00:00:00", scale="tdb")
```

```python
#NAS/JPLよりdidymosの軌道データを呼び出して位置を計算させる
didymos = Ephem.from_horizons("65803", EPOCH, plane=Planes.EARTH_ECLIPTIC)
didymos

#NAS/JPLよりdidymosの軌道データを呼び出して軌道を計算させる
didymos_orbit= Orbit.from_sbdb("65803")
```

```python
#グラフを書く
frame = StaticOrbitPlotter(plane=Planes.EARTH_ECLIPTIC)
#惑星を描写
frame.plot_body_orbit(Earth, EPOCH, label="Earth")
frame.plot_body_orbit(Mars, EPOCH,label="Mars")
frame.plot_body_orbit(Jupiter, EPOCH, label="Jupiter")
#didymosを描写する
didymos_orbit_plot_t,didymos_orbit_plot_p=frame.plot(didymos_orbit)
didymos_orbit_plot_p.set_marker("")
frame.plot_ephem(didymos, EPOCH, label="didymos")
```
